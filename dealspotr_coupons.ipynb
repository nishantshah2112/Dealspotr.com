```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Setup Selenium WebDriver
service = Service(
    r"C:\Users\Nishant shah\OneDrive\Desktop\Nishant\Software\chromedriver.exe"
)
driver = webdriver.Chrome(service=service)

# URLs to scrape
urls = [
    "https://dealspotr.com/promo-codes/search/cat=devices",
    "https://dealspotr.com/promo-codes/search/cat=health",
    "https://dealspotr.com/promo-codes/search/cat=beauty",
    "https://dealspotr.com/promo-codes/search/cat=fashion",
]

all_data = []

try:
    # Iterate through each URL
    for url in urls:
        print(f"Scraping URL: {url}")
        driver.get(url)
        time.sleep(3)

        # Loop to click "View More" until the last page
        while True:
            try:
                view_more_button = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable(
                        (By.XPATH, "//div[contains(@class, 'exp--button--wide')]")
                    )
                )
                view_more_button.click()
                print("Clicked 'View More'. Waiting for content to load...")
                time.sleep(3)  # Wait for new content to load
            except Exception as e:
                print("No more 'View More' button found or an error occurred:", e)
                break

        # Extract product blocks
        product_blocks = driver.find_elements(By.CLASS_NAME, "flex2--module")

        for product in product_blocks:
            try:
                # Extract the promo block (URL)
                promo_block = product.find_element(By.CLASS_NAME, "promoblock--logo")
                data_url = promo_block.find_element(By.TAG_NAME, "img").get_attribute(
                    "data-url"
                )

                # Extract the coupon code
                try:
                    coupon_element = product.find_element(By.CLASS_NAME, "dnone")
                    coupon_code = coupon_element.get_attribute("value")
                except Exception as e:
                    coupon_code = "N/A"
                    print("Coupon Code Error:", e)

                # Append the data
                all_data.append({"URL": data_url, "Coupon Code": coupon_code})
                print(f"Scraped Data: URL={data_url}, Coupon Code={coupon_code}")

            except Exception as e:
                print("Error while scraping product data:", e)
                continue

finally:
    # Save the data to an Excel file
    df = pd.DataFrame(all_data)
    df.to_excel("dealspotr_coupons.xlsx", index=False, engine="openpyxl")
    print("Data saved to 'dealspotr_coupons.xlsx'.")

    # Close the browser
    driver.quit()
```
